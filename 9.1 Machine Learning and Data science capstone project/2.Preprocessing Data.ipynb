```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv("cisco_firewall_final_dataset_with_severity.csv")
```

```python
print("Dataset Shape:", df.shape)
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df_prep = df.copy()
```

```python
# Encode severity
severity_encoder = LabelEncoder()
df_prep["severity_encoded"] = severity_encoder.fit_transform(df_prep["severity_level"])
```

```python
# One-hot encode model_sku
df_prep = pd.get_dummies(df_prep, columns=["model_sku"], drop_first=True)
```

```python

# Feature scaling
scale_cols = [
    "cpu_util_avg","memory_util_avg","port_util_avg","interface_error_rate",
    "connection_rate_cps","syn_flood_events","inspection_drop_rate",
    "fragmented_packet_pct","image_upgrade_count","reboot_count","failover_events",
    "core_file_count","device_stability_score"
]

scaler = StandardScaler()
df_prep[scale_cols] = scaler.fit_transform(df_prep[scale_cols])
```

```python
# Drop original severity
df_prep.drop(columns=["severity_level"], inplace=True)
```

```python
# Save preprocessed dataset
df_prep.to_csv("cisco_firewall_preprocessed_dataset.csv", index=False)
df_prep
```
